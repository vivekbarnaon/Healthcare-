```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    dia.drop(columns=['diabetes']),
    dia['diabetes'],
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
transformer = ColumnTransformer(
    transformers=[
        ('tnf1', OrdinalEncoder(categories=[['never', 'former', 'current', 'not current', 'ever', 'No Info']]), ['smoking_history']),
        ('tnf2', OneHotEncoder(), ['gender'])
    ],
    remainder='passthrough'
)
```

```python
t = transformer.fit_transform(X_train)
t_ = transformer.transform(X_test)
```

```python
t_
```

```python
X_train
```

```python
encoded_columns = ['smoking_history_encoded']  # 1 column for smoking_history
gender_columns = ['gender_female', 'gender_male']  # 2 columns for gender
remaining_columns = [col for col in X_train.columns if col not in ['smoking_history', 'gender']]
new_column_names = encoded_columns + gender_columns + remaining_columns
```

```python
transformed_columns = transformer.get_feature_names_out(input_features=X_train.columns)
```

```python
new_column_names
```

```python
X_train_transformed = pd.DataFrame(t, columns=transformed_columns, index=X_train.index)
X_test_transformed = pd.DataFrame(t_, columns=transformed_columns, index=X_test.index)
```

```python
X_train_transformed
```

```python
X_train_transformed.columns
```

```python
new_column_names = [col.split('__')[-1] for col in X_train_transformed.columns]

X_train_transformed.columns = new_column_names
X_test_transformed.columns = new_column_names
```

```python
X_train_transformed
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

model = LogisticRegression(max_iter=1000)
model.fit(X_train_transformed, y_train)

y_pred = model.predict(X_test_transformed)

accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(model, X_train_transformed, y_train, cv=5)
print(f"Cross-Validation Scores: {cv_scores}")
print(f"Mean Cross-Validation Score: {cv_scores.mean()}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train_transformed, y_train)

# Predict on the test data
y_pred = rf_model.predict(X_test_transformed)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import classification_report

# Initialize the LightGBM model with class imbalance handling
model = LGBMClassifier(is_unbalance=True)

# Train the model on the transformed data
model.fit(X_train_transformed, y_train)

# Predict on the test set
y_pred = model.predict(X_test_transformed)

# Evaluate the model
print("Accuracy:", model.score(X_test_transformed, y_test))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from lightgbm import LGBMClassifier

# Define the model with class_weight
model = LGBMClassifier(class_weight={0: 1, 1: 10}, is_unbalance=True)

# Train and evaluate the model
model.fit(X_train_transformed, y_train)
y_pred = model.predict(X_test_transformed)

# Evaluate the model
print("Accuracy:", model.score(X_test_transformed, y_test))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE

# Apply SMOTE to balance the dataset
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_transformed, y_train)

# Train the model
model = LGBMClassifier()
model.fit(X_train_smote, y_train_smote)

# Get predicted probabilities
y_prob = model.predict_proba(X_test_transformed)[:, 1]

# Set a new threshold (e.g., 0.4 instead of 0.3)
y_pred_new = (y_prob >= 0.4).astype(int)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred_new))
print("Classification Report:")
print(classification_report(y_test, y_pred_new))
```
